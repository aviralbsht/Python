```python
#Shapes area calculator
shape = str(input("Select shape Rectangle,Square,Triangle,Circle"))
if shape == "Rectangle":
  length = int(input("Enter the length"))
  width = int(input("Enter the width"))
  rect_area = length*width
  print(f"area of Rectangle is:{rect_area}")
elif shape == "Square":
  length = int(input("Enter the length"))
  sqr_area = length*length
  print(f"area of Square is: {sqr_area}")
elif shape == "Triangle":
  hieght = int(input("Enter the hieght"))
  base = int(input("Enter the base"))
  tri_area = 0.5*hieght*base
  print(f"area of Triangle is: {tri_area}")
elif shape == "circle":
  radius = int(input("Enter the radius length"))
  pi = 3.14
  circle_area = pi*radius*radius
  print(f"area of circle is:{circle_area}")
else:
  print("This shape is not available")
  exit()
```

```python
# Python calculator
operator = input("Select an Operator from (+,-,*,/)")
num1 = float(input("Enter 1st No."))
num2 = float(input("Enter 2nd No."))
if operator == "+":
  result = num1 + num2
  print(result)
elif operator == "-":
  result = num1 - num2
  print(result)
elif operator == "*":
  result = num1 * num2
  print(result)
elif operator == "/":
  result = num1 / num2
  print(result)
else:
  print("Operator not available")
  exit()
```
